```python
def plot_polar(ax, phi_list, theta_list, r0, color='k'):
    phi = np.array(phi_list)
    theta = np.array(theta_list)
    
    phi_deg = phi * 180/np.pi
    wedge_list = []
    for i in range(0, len(phi)-1):
        wedge_list.append(
            Wedge(\
                (0,0), \
                theta[i] + r0, \
                -90-phi_deg[i+1], \
                -90-phi_deg[i]\
            ) \
        )
    
    p = PatchCollection(wedge_list, color=color)
    ax.add_collection(p)
```

```python
# polar plot
r0 = 2

fig = plt.figure()
ax = fig.add_subplot(111)

# plot sun
x_list = []
y_list = []
for p in sun.positions:
    r = r0 + p[1]
    x_list.append(-r * np.sin(p[0]))
    y_list.append(-r * np.cos(p[0]))
    
ax.plot(x_list, y_list, color='#ffa700', zorder = 0, linewidth=3.0)

# plot skyline
phis = []
thetas = []
for cliff in sil.cliffs:
    phis.append(cliff.phi)
    thetas.append(cliff.theta_L) 
    phis.append(cliff.phi)
    thetas.append(cliff.theta_R) 
plot_polar(ax, phis, thetas, r0)

# # plot horizon
horizon = Circle((0,0), r0)
h = PatchCollection([horizon], color='w')
ax.add_collection(h)

# size of plot
L = r0 + np.pi/2
d = 0.1


# geographic directions
gray_color = '#909090'
ax.plot([0,0], [-L,L], color=gray_color, zorder=10)
ax.plot([-L,L], [0,0], color=gray_color, zorder=10)

ax.text(0, -L, 'South', verticalalignment=u'bottom', horizontalalignment=u'center')
ax.text(0, L, 'North', verticalalignment=u'top', horizontalalignment=u'center')
ax.text(-L, 0, 'West', verticalalignment=u'bottom', horizontalalignment=u'left')
ax.text(L, 0, 'East', verticalalignment=u'bottom', horizontalalignment=u'right')

# axes
ax.set_ylim([-L, L])
ax.set_xlim([-L, L])
ax.xaxis.set_visible(False)
ax.yaxis.set_visible(False)


ax.set_aspect('equal')
fig.set_size_inches(8, 8)

```

```python
def plot_fisheye(ax, phi_list, theta_list):
    phi = np.array(phi_list)
    theta = np.array(theta_list)
    
    
    phi_deg = phi * 180/np.pi
    for i in range(0, len(phi)):
        if phi_deg[i] < -180:
            phi_deg[i] = -180
        if phi_deg[i] > 180:
            phi_deg[i] = 180
    wedge_list = []
    for i in range(0, len(phi)-1):
        wedge_list.append(
            Wedge(\
                (0,0), \
                np.pi/2 - theta_list[i], \
                -90+phi_deg[i], \
                -90+phi_deg[i+1]
            ) \
        )
    
    p = PatchCollection(wedge_list, color='w')
    ax.add_collection(p)
```
