```python
%matplotlib inline
import re
import csv
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np
import datetime as dt
import pymysql as mdb
```

```python
class Node:
    def __init__(self, input_x=0, input_y=0):
        self.x = input_x
        self.y = input_y
    
    def show(self):
        print '[x=' + str(self.x) + ', y=' + str(self.y) + ']'
```

```python
# class to store geometries of a single building
class Building:
    def __init__(self):
        self.nodes = []   # list of Nodes
        self.center = Node()
        self.z = None        # height
        
    def show(self):
        print 'Nodes:'
        for node in self.nodes:
            node.show()
        print 'Center:\n' + 'x_center=' + str(self.center.x) + ', ' \
              + 'y_center=' + str(self.center.y)
        print 'Height:\n' + 'z=' + str(self.z)
        
    def plot_footprint(self):
        x_list = []
        y_list = []
        for node in self.nodes:
            x_list.append(node.x)
            y_list.append(node.y)
        plt.plot(x_list,y_list)
        
    def calculate_center(self):
        x_array = []
        y_array = []
        for node in self.nodes:
            x_array.append(node.x)
            y_array.append(node.y)
        self.center.x = np.mean(x_array)
        self.center.y = np.mean(y_array)
```

```python
class Block:
    def __init__(self):
        self.vertices = []  # list of nodes
        self.building_ids = []  # list of building_ids
        self.max_z = 0  # maximum height of the buildings in it
    
    def __init__(self, node1, node2, node3, node4):
        self.vertices = [node1, node2, node3, node4]  # list of nodes
        self.building_ids = []  # list of building_ids
        self.max_z = 0  # maximum height of the buildings in it
    
    def calculate_max_z(self, buildings):
        current_max_z = 0
        for key in building_ids:
            this_z = builgings[key].z
            if this_z > current_max_z:
                current_max_z = this_z
        self.max_z = current_max_z
    
    def plot_on_map(self):
        x_list = []
        y_list = []
        for node in self.vertices:
            x_list.append(node.x)
            y_list.append(node.y)
        plt.plot(x_list, y_list)
```

```python
def is_outside(boundary_node_1, boundary_node_2, building_node):
    ux = building_node.x - boundary_node_1.x
    uy = building_node.y - boundary_node_1.y
    vx = boundary_node_2.x - boundary_node_1.x
    vy = boundary_node_2.y - boundary_node_1.y
    
    if ((vx*uy) > (vy*ux)):
        return True
    else:
        return False
```

```python
def assign_to_block(building_id, building, x_grid, y_grid, blocks):
    
    # find x_index of the block
    block_x_index = -1
    for x in x_grid:
        if building.center.x > x:
            block_x_index += 1
        else:
            break
    
    # find y_index of the block
    block_y_index = -1
    for y in y_grid:
        if building.center.y > y:
            block_y_index += 1
        else:
            break
            
    # compile key (as string)
    block_key = str(block_x_index) + ':' + str(block_y_index)
    
    # add the building id to the selected block
    blocks[block_key].building_ids.append(building_id)
    
    # add the selected block's id to the building
    building.block_id = block_key
        
```

```python
# choose file and load it in raw
#data_directory = '~/InsightDataFellowship/Project/Data/'
data_directory = '../Data/'
xml_filename = data_directory + 'Empire.xml'
file = open(xml_filename, 'r')
filecontent = file.read()
file.close()


# collect the nodes into a dict
# empty dict for nodes {id: Node}
nodes = {}
node_match_str = r'<node id="(\d+)" lat="(-?\d+\.\d+)" lon="(-?\d+\.\d+)"/?>'
match = re.findall(node_match_str, filecontent)
for node in match:
    nodes[str(node[0])] = Node() # id is the numer 0 output
    nodes[str(node[0])] .x = float(node[2]) # lon is the number 2 output
    nodes[str(node[0])] .y = float(node[1]) # lat is the number 1 output

    
# collect buildings into dict
buildings = {}

# first, find ways
way_match_str = r'<way id="(\d+)">([\w\W]+?)</way>'
matched_ways = re.findall(way_match_str, filecontent)
# match[i][0] : ith way id
# match[i][1] : attributes of ith way

for way in matched_ways:
    # get building id
    found_id = way[0]
    buildings[found_id] = Building()
    
    # get list of node ids
    found_attributes = way[1]
    node_mach_str = r'<nd ref="(\d+)"/>'
    matched_nodes = re.findall(node_mach_str, found_attributes)
    
    for node in matched_nodes:
        buildings[found_id].nodes.append(nodes[node])
        
    # calculate center
    buildings[found_id].calculate_center()
    
    # get height
    height_ref_match = r'<tag k="height" v="(\d+\.?\d*)"/>'
    matched_height = re.findall(height_ref_match, found_attributes)
    if matched_height:
        buildings[found_id].z = float(matched_height[0])
    
# drop buildings with no height
no_height_keys = []
for key in buildings:
    if buildings[key].z is None:
        no_height_keys.append(key)
for key in no_height_keys:
    del buildings[key]
```

```python
# let's convert the (lon,lat) coordinates into (x[m], y[m])

# mean radius of Earth (in meters)
R = 6371009

# mean coordinates of Manhattan
mean_lat =  40.771803
mean_lon = -73.973351
deg2rad = np.pi/180
cos_mean_lat = np.cos(mean_lat * deg2rad)

buildings_cartesian = {}
for key in buildings:
    buildings_cartesian[key] = Building()
    for node in buildings[key].nodes:
        this_x = R * cos_mean_lat * (node.x - mean_lon)* deg2rad
        this_y = R * (node.y - mean_lat)* deg2rad
        this_node = Node(this_x, this_y)
        buildings_cartesian[key].nodes.append(this_node)
    buildings_cartesian[key].calculate_center()
    buildings_cartesian[key].z = buildings[key].z
```

```python
# definition of the gridlines for grouping into blocks
x_grid = np.arange(-5e3,+7e3, 1e3)
y_grid = np.arange(-1e4, 1.4e4, 1e3)

# initialize the blocks
blocks = {}
for j in range(0, len(y_grid)-1):
    for i in range(0, len(x_grid)-1):
        node1 = Node(x_grid[i], y_grid[j])
        node2 = Node(x_grid[i+1], y_grid[j])
        node3 = Node(x_grid[i+1], y_grid[j+1])
        node4 = Node(x_grid[i], y_grid[j+1])
        block_id = str(i) + ':' + str(j)
        blocks[block_id] = Block(node1, node2, node3, node4)

# assign each building to one block
for key in buildings_cartesian:
    assign_to_block(key, buildings_cartesian[key], x_grid, y_grid, blocks)

# drop blocks that are empty
empty_blocks = []
for key in blocks:
    if not blocks[key].building_ids:
        empty_blocks.append(key)
for key in empty_blocks:
    del blocks[key]
```
