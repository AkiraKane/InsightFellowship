```python
%matplotlib inline
import pymysql as mdb

# my modules
from buildingmapping import *
from skyline import *
from sun import *
from observer import *
```

```python
obs = Observer(lat=40.765697, lon=-73.976232)
buildings = {}
sil = Silhouette()
summary = SunSummary()
```

```python
con = mdb.connect('localhost', 'root', '123', 'Manhattan_buildings')
obs.load_basic_geography(con)
(x_grid, y_grid) = load_grid_data(con)
obs.convert_to_cartesian()
obs.find_my_block(x_grid, y_grid)
buildings.clear()
(x_id_list, y_id_list) = obs.get_neighboring_block_ids()
for i in range(0, len(x_id_list)):
    buildings.update(\
        append_buildings_in_block(con, x_id_list[i], y_id_list[i])\
    )
sil.cliffs = Silhouette().cliffs
for key in buildings:
    if buildings[key].z > obs.z:
        roofs = buildings[key].get_roofs(obs.x, obs.y, obs.z)
        for tup in roofs:
            sil.add_roof( Roof(tup) )
```

```python
summary.collect_summary(sil, obs)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])


color_total = '#ffc100'
color_morning = '#ff9a00'
color_afternoon = '#ff5a00'
color_sunset = '#b60505'
dashed_style = ':'
height_sunset_line = 500
height_of_plot = 600

week_list = np.arange(1, len(summary.dates)+1,1)
L = len(week_list)

dates = summary.dates
total = np.array(summary.total_sun)
morning = np.array(summary.morning_sun)
afternoon = np.array(summary.afternoon_sun)
sunrise = summary.sunrise
sunset = summary.sunset

# plot curves for the minutes of sun
ax.plot(week_list, total/2, color='k', linestyle=dashed_style)
ax.plot(week_list , -total/2, color='k', linestyle=dashed_style)
ax.fill_between(week_list , 0, - morning , color=color_morning)
ax.fill_between(week_list , 0, afternoon, color=color_afternoon)
ax.set_xlim([1, len(summary.dates)])  
    
    
    
# # plot sunset/sunrise

# # find sunrises
# i = 0
# while i < L:
#     if not sunrise[i]:
#         i += 1
#     else:
#         x_list = []
#         y_list = []
#         while sunrise[i]: 
#             x_list.append(week_list[i])
#             y_list.append(-height_sunset_line)
#             i += 1
#             if not i < L:
#                 break
#         ax.plot(x_list, y_list, color=color_sunset, linewidth=2)
        
#     i += 1

# # find sunsets
# i = 0
# while i < L:
#     if not sunset[i]:
#         i += 1
#     else:
#         x_list = []
#         y_list = []
#         while sunset[i]: 
#             x_list.append(week_list[i])
#             y_list.append(height_sunset_line)
#             i += 1
#             if not i < L:
#                 break
#         ax.plot(x_list, y_list, color=color_sunset, linewidth=4)

#     i += 1

    

first_weeks = []

d = dates[0]
year = d.year
one_day = dt.timedelta(days=1)
current_month = 0
i = 0
while d.year == year :
    if d.month > current_month:
        first_weeks.append(1 + i / 7.0)
        current_month += 1
    d += dt.timedelta(days=1)
    i += 1

for i in first_weeks:
    ax.plot([i,i], [height_of_plot, -height_of_plot], color='k', linewidth=0.5)

ax.set_ylabel('minutes of morning/afternoon sun')
    

month_names = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
for i in range(0, 12, 1):
    ax.text(first_weeks[i]+1.5, height_of_plot - 50, month_names[i])
    ax.text(first_weeks[i]+1.5, -height_of_plot + 30, month_names[i])

# x_ticks = range(1, L+1, 1)
# plt.xticks(x_ticks)

y_ticks = range(-600, +700, 100)
y_labels = [str(np.abs(yt)) for yt in y_ticks]
plt.yticks(y_ticks, y_labels)

ax.xaxis.set_visible(False)

fig.set_size_inches(8, 4)
plt.show()

```
